```python
## create dataframe to store items


tweet_db = pd.DataFrame(columns = ['id','text','date','hashtags', 'rt_count', 'rt', 'source'])
for tweet in tweets:
    
    #### convert hashtags to list
    hts = []   #make an empty list
    for hs in tweet.entities["hashtags"]:    #iterate over the list
        hts.append(hs["text"])
        
        
    ## add row to table
    tweet_db = tweet_db.append({
        'id': tweet.id,
        'text': tweet.full_text,
        'date': tweet.created_at,
        'hashtags': ''.join(hts),   ### convert list to string
        'rt_count': tweet.retweet_count, 
        'rt': tweet.retweeted, 
        'source': tweet.source      
    }, ignore_index=True)

tweet_db
```

```python
### write function to convert tweets to numbers
def tweets_to_df(tweets):
    tweet_db = pd.DataFrame(columns = ['id','text','date','hashtags', 'rt_count', 'rt', 'source'])
    for tweet in tweets:

        #### convert hashtags to list
        hts = []   #make an empty list
        for hs in tweet.entities["hashtags"]:    #iterate over the list
            hts.append(hs["text"])


        ## add row to table
        tweet_db = tweet_db.append({
            'id': tweet.id,
            'text': tweet.full_text,
            'date': tweet.created_at,
            'hashtags': ''.join(hts),   ### convert list to string
            'rt_count': tweet.retweet_count, 
            'rt': tweet.retweeted, 
            'source': tweet.source      
        }, ignore_index=True)

    return tweet_db
```

```python
import matplotlib.pyplot as plt

tweet_db['id'].groupby(tweet_db["date"].dt.date).count().plot(kind="bar")
```

```python
###get the next group, maximum is 200, default is 20
new_tweets_extend = api.user_timeline(screen_name = 'SpaceX',count=20, max_id=tweet_db['id'].min()-1, tweet_mode = 'extended')

#### check length of new pull
extend = tweets_to_df(new_tweets_extend)
extend
```

```python
comb = pd.concat([tweet_db, extend])
comb.reset_index(inplace = True, drop=True)
comb
```

```python
comb['id'].groupby(comb["date"].dt.date).count().plot(kind="bar")
```

```python
###start with first pull
###################code to pull top 3000 tweets for the user


name = 'ladygaga'
new_tweets = api.user_timeline(screen_name = name,count=200, tweet_mode = 'extended')
### establish db
full_tweet = pd.DataFrame(columns = ['id','text','date','hashtags', 'rt_count', 'rt', 'source'])

while (len(new_tweets)>0):   ###while there are more tweets to read
    
    for tweet in new_tweets:   ##for each tweet
    
        #### convert hashtags to list
        hts = []   #make an empty list
        for hs in tweet.entities["hashtags"]:    #iterate over the list
            hts.append(hs["text"])
        
        
        ## add row to table
        full_tweet = full_tweet.append({
            'id': tweet.id,
            'text': tweet.full_text,
            'date': tweet.created_at,
            'hashtags': ''.join(hts),   ### convert list to string
            'rt_count': tweet.retweet_count, 
            'rt': tweet.retweeted, 
            'source': tweet.source      
        }, ignore_index=True)
    
    #after finishing the previous batch, get a new batch
    new_tweets = api.user_timeline(screen_name = name,count=200, max_id=full_tweet['id'].min()-1, tweet_mode = 'extended')
    
full_tweet
```

```python
#### use search term

search_db = pd.DataFrame(columns = ['id','user','text','date','hashtags', 'rt_count', 'rt', 'source'])

for tweet in tweepy.Cursor(api.search_tweets, 
                    q="Rutgers", 
                    result_type='popular',
                    tweet_mode = 'extended',
                    lang="en"
                          ).items(1000):
     #### convert hashtags to list
        hts = []   #make an empty list
        for hs in tweet.entities["hashtags"]:    #iterate over the list
            hts.append(hs["text"])
        
        
        ## add row to table
        search_db = search_db.append({
            'id': tweet.id,
            'user':tweet.user.name,
            'text': tweet.full_text,
            'date': tweet.created_at,
            'hashtags': ' '.join(hts),   ### convert list to string
            'rt_count': tweet.retweet_count, 
            'rt': tweet.retweeted, 
            'source': tweet.source      
        }, ignore_index=True)
        
search_db
```
