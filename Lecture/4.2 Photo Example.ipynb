# Photo Example

```python
import numpy as np
from skimage import io
```

```python
photo = io.imread('RutgersWinter.jpg')
```

```python
## Check type of photo
type(photo)
```

```python
photo.shape
##width, length, and 3 because RGB color
```

```python
##Take a look at photo
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
### let's go backward
plt.imshow(photo[::-1, :])
```

```python
### let's go backward
plt.imshow(photo[::3, ::3])
```

```python
### take a slice of the photo
plt.imshow(photo[30:240, 100:430])
```

```python
### mask the photo
plt.imshow(np.where(photo > 100, 255,0))
```

```python
### doing transpose
plt.imshow(photo[:,:,0].T)
```
